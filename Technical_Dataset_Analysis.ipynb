```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Technical_Dataset_500_Rows.xlsx"  # Replace with the correct path
data = pd.read_excel(file_path)

# Display the first few rows of the dataset
print("Dataset Preview:")
print(data.head())

# Summary statistics of the dataset
print("\nSummary Statistics:")
print(data.describe())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Basic information about the dataset
print("\nDataset Info:")
print(data.info())


# Correlation matrix (exclude non-numerical columns)
numeric_columns = data.select_dtypes(include=[np.number])  # Select only numeric columns
correlation_matrix = numeric_columns.corr()

print("\nCorrelation Matrix (Numeric Columns):")
print(correlation_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap (Numeric Columns)")
plt.show()
```

```python
# Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(data['Age'], kde=True, bins=15, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Average monthly salary by department
avg_salary = data.groupby('Department')['Monthly_Salary'].mean().sort_values()
print("\nAverage Monthly Salary by Department:")
print(avg_salary)
```

```python
plt.figure(figsize=(8, 5))
avg_salary.plot(kind='bar', color='orange')
plt.title("Average Monthly Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Monthly Salary")
plt.show()
```

```python
# Boxplot of Monthly Salary by Department
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Department', y='Monthly_Salary', palette='pastel')
plt.title("Monthly Salary Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Monthly Salary")
plt.show()
```

```python
# Satisfaction score vs Performance score
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Satisfaction_Score', y='Performance_Score', hue='Resigned', palette='viridis')
plt.title("Satisfaction Score vs Performance Score")
plt.xlabel("Satisfaction Score")
plt.ylabel("Performance Score")
plt.show()
```

```python
# Resignation count by department
resignation_counts = data[data['Resigned'] == 'Yes']['Department'].value_counts()
print("\nResignation Count by Department:")
print(resignation_counts)
```

```python
plt.figure(figsize=(8, 5))
resignation_counts.plot(kind='bar', color='red')
plt.title("Resignation Count by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
# Top performers (Performance Score > 4.5)
top_performers = data[data['Performance_Score'] > 4.5]
print("\nTop Performers (Performance Score > 4.5):")
print(top_performers)
```

```python
# Export top performers to a new Excel file
top_performers.to_excel("Top_Performers.xlsx", index=False)
print("\nTop performers data exported to 'Top_Performers.xlsx'.")
```
